```python
#Reading the data
#bids.csv

bids = pd.read_csv(r'C:\Users\cinti\Documents\PythonF\Final_Project_ML\bids.csv', sep=',')
print('bids: OK {}'.format(bids.shape))
bids.head()
```

```python
#Reading the data
#train.csv

b_train = pd.read_csv(r'C:\Users\cinti\Documents\PythonF\Final_Project_ML\train.csv', sep=',')
print('b_train: OK: {}'.format(b_train.shape))
b_train.head()
```

```python
#Include the information from the test data to the bids - get if the outcome is bot, human or not available 

bids['outcome'] = bids.merge(b_train, how='left', on='bidder_id').outcome
```

```python
bids.head()
```

```python
bids.shape
```

```python
sum_bots = bids.loc[(bids['outcome'] == 0)].count()
print(sum_bots)
```

```python
sum_human = bids.loc[(bids['outcome'] == 1)].count()
print(sum_human)
```

```python
sum_none = bids['outcome'].isnull().sum()
print(sum_none)
```

```python
bids = bids[bids['outcome'].notna()]
```

```python
bids.shape
```

```python
print('Data types:')
print(bids.dtypes)
```

```python
bids['bid_id'] = bids['bid_id'].astype('category')
bids['bidder_id'] = bids['bidder_id'].astype('category')
bids['auction'] = bids['auction'].astype('category')
bids['merchandise'] = bids['merchandise'].astype('category')
bids['device'] = bids['device'].astype('category')
bids['country'] = bids['country'].astype('category')
bids['ip'] = bids['ip'].astype('category')
bids['url'] = bids['url'].astype('category')
bids['outcome'] = bids['outcome'].astype('category')

bids.dtypes
```

```python
# Missing values

print(bids.isnull().sum())
```

```python
# Country is the one with missing information 

print(bids['country'].value_counts())
```

```python
# Use Most commom occurences to fill null values in country 

bids['country'] = bids['country'].fillna('in')

print(bids.isnull().sum())
```

```python
# Verify the number of bids made by Robots and number of bids made by humans in the dataset 

made_by_robots = len(bids[bids['outcome'] == 1])
made_by_humans = len(bids[bids['outcome'] == 0])

print('Number of bids made by robots:', made_by_robots)
print('Number of bids made by humans:', made_by_humans)
```

```python
inf_bidders = bids.filter(['bidder_id'], axis = 1)

inf_bidders
```

```python
inf_bidders.shape
```

```python
i = 0
for bidder in inf_bidders['bidder_id']:
    if i % 100 == 0:
        print(i)
    i += 1
    c = bids[bids.bidder_id == bidder].count()[0]
    inf_bidders.loc[inf_bidders[inf_bidders.bidder_id == bidder].index, 'total_bids'] = c
```
